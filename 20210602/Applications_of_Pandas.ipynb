## 3 維陣列的運用

```python
# arr_image 中元素的型態 (註: uint 表示無正負號的整數)
type_ = arr_image.dtype  

# arr_image 的維度(Dimension)，或稱軸(Axis)。
dim_ = arr_image.ndim

# arr_image 的元素個數
size_ = arr_image.size

# arr_image 的形狀，表示每個維度的元素個數。
shape_ = arr_image.shape

print( 'Type of Element in Array :', type_ )
print( 'Dimension of Array =', dim_ )
print( 'Size of Array =', size_ )
print( 'Shape of Array =', shape_ )
```

# DataFrame  

DataFrame 為具備列(row)及欄(column)索引的表格物件，它的功用如同 Excel 試算表或 SQL 資料表，可對它進行篩選、排序、樞紐分析及統計、新增或刪減欄位等。 

- [`pd.DataFrame()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html) 函數可用來建立一個 DataFrame；
- `shape` 屬性可得知 DataFrame 的列數與欄數；  
- `info()` 屬性可得知 DataFrame 中每個欄位的型態及非缺失的資料總數；
- [`pandas.DataFrame.describe()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.describe.html) 函數可得知數值型態欄位的分佈狀況；  
- [`pandas.DataFrame.loc[ ]`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html) 可藉由列/行的標籤(列索引名稱/欄位名稱)來取得資料；  
- [`pandas.DataFrame.iloc[ ]`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html) 可藉由列/行的索引值(列索引/欄索引)來取得資料；  
- [`pandas.DataFrame.apply()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html) 搭配 [lambda](https://www.w3schools.com/python/python_lambda.asp) 匿名運算式，可用來操作 DataFrame 欄位之間的四則與邏輯運算。

### 統計函式  

 - [`pandas.DataFrame.count(axis=0)`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.count.html) : 計算欄位中非空的資料總數；
 - [`pandas.DataFrame.sum(axis=None)`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sum.html) : 計算列/欄中的總和；
 - [`pandas.DataFrame.prod(axis=None)`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sum.html) : 計算列/欄中的乘積； 
 - [`pandas.DataFrame.mean(axis=None)`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.mean.html) : 計算列/欄中的平均值； 
 - [`pandas.DataFrame.median(axis=None)`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.median.html) : 計算列/欄中的中位數； 
 - [`pandas.DataFrame.std(axis=None)`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.std.html) : 計算列/欄中的標準差； 
 - [`pandas.DataFrame.max(axis=None)`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.max.html) : 求得列/欄中的最大值； 
 - [`pandas.DataFrame.min(axis=None)`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.min.html) : 求得列/欄中的最小值。

## 排序資料

[`sort_values( by, ascending=True, ignore_index=False, inplace=False )`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html) 函數可對指定的欄位進行排序，其中， `by` 參數為指定的欄位名稱， `ascending` 參數為排序的方式(預設為 True 由小到大)：

```python
dict_ = { 'Name': ['Steven','Bill','Joe','Andy','Tom'],
          'Height': [172,179,170,165,172],
          'Weight': [82,95,65,76,70] }

df = pd.DataFrame( dict_ )
df
```

```python
# 由小到大排序 Height 欄位
sorted_height = df.sort_values( 'Height' )
sorted_height
```

```python
# 由小到大先排序 Height 欄位，再排序 Weight 欄位。
sorted_df = df.sort_values( ['Height','Weight'] )
sorted_df
```

```python
# 由小到大先排序 Height 欄位，再排序 Weight 欄位，且重新定義索引值(index)。
sorted_df = df.sort_values( ['Height','Weight'], ignore_index=True )
sorted_df
```

## 整合多個資料

  * `other` : 欲連接的第二個資料，可為 DataFrame 或 Series 物件 ；  
  * `ignore_index` : 是否重新排序索引值( index )，預設值為 False 。

## 樞紐分析

在 Excel 中，樞紐分析表係一個能從原始資料中迅速歸納計算出統計報表的工具。在 Python 中，我們可使用   
 [`pandas.pivot_table( data, index, columns, values, aggfunc='mean', dropna=True, margins=False )`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html)   
函數對指定的 DataFrame 物件進行樞紐分析，參數說明如下：    
  * `data` : DataFrame物件 ；  
  * `index` : 樞紐分析表中的列 ；
  * `columns` : 樞紐分析表中的欄 ；
  * `values` : 計算樞紐分析表中值的 DataFrame 欄位 ；  
  * `aggfunc` : 計算樞紐分析表中值的方式，預設為平均值('mean') ； 
  * `dropna` : 計算樞紐分析表中值的過程中，是否排除空值，預設為 True ； 
  * `margins` : 樞紐分析表中是否需要總計(或小計)，預設為 False 。  

```python
df = pd.DataFrame( {'Product': ['A','A','A','A','A','B','B','B','B'],
                    'Size': ['Big','Big','Small','Small','Small','Big','Big','Small','Small'],
                    'Amount': [10,10,20,30,np.nan,60,70,80,70],
                    'Revenue': [1500,2000,500,100,np.nan,4500,5000,150,100]} )

df
```

```python
# 依 'Product' 與 'Size' 欄位做分組，計算出每一組 'Amount' 的總和，且樞紐分析表的欄位為 'Size' 欄位中的類別資料。
pivot_table = pd.pivot_table( df, index='Product', columns='Size', values='Amount', aggfunc=np.sum )
pivot_table
```

```python
# 依 'Product' 與 'Size' 欄位做分組，計算出每一組 'Amount' 的總和。
pivot_table = pd.pivot_table( df, index=['Product','Size'], values='Amount', aggfunc=np.sum )
pivot_table
```

```python
# 將樞紐分析表的 index 欄位轉換成 DataFrame 的欄位
pivot_table.reset_index( inplace=True )
pivot_table
```

```python
# 依 'Product' 與 'Size' 欄位做分組，計算出每一組 'Amount' 的總和與 'Revenue' 的平均。
pivot_table = pd.pivot_table( df, index=['Product','Size'], values={'Amount','Revenue'}, aggfunc={'Amount':np.sum,'Revenue':np.mean} )
pivot_table
```

另外一種方法，我們也可使用 [`pandas.DataFrame.groupby()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html) 函數進行樞紐分析 : 

```python
# 依 'Product' 與 'Size' 欄位做分組，計算出每一組 'Amount' 的總和。
pivot_table = df[['Product','Size','Amount']].groupby(['Product','Size']).sum()
pivot_table
```

```python
# 依 'Product' 與 'Size' 欄位做分組，計算出每一組 'Amount' 與 'Revenue' 的總和。
pivot_table = df.groupby(['Product','Size']).sum()
pivot_table
```

另外，透過 [`pandas.DataFrame.agg()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.agg.html) 函數可針對特定的欄位進行指定的統計方式 : 
  
參考資料 :  
 - [Get statistics for each group (such as count, mean, etc) using pandas GroupBy?](https://stackoverflow.com/questions/19384532/get-statistics-for-each-group-such-as-count-mean-etc-using-pandas-groupby)
 - [Applying multiple functions at once](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html#applying-multiple-functions-at-once)

```python
# 依 'Product' 與 'Size' 欄位做分組，計算出每一組 'Amount' 的總和與 'Revenue' 的平均。
pivot_table = df.groupby(['Product','Size']).agg( {'Amount':'sum','Revenue':'mean'} )
pivot_table
```
